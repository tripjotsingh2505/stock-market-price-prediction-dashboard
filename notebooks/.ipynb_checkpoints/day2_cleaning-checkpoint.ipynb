```python
import pandas as pd
import numpy as np
import yfinance as yf

# Thoda alag tarike se download karo
ticker = yf.Ticker("RELIANCE.NS")
df = ticker.history(start="2020-01-01", end="2026-04-05")

print("Data loaded successfully! ✅")
print("Shape:", df.shape)
df.head()
```

```python
df.tail()
```

```python
# Missing values check karo
print("Missing values har column mein:")
print(df.isnull().sum())
```

```python
# Duplicate rows check karo
print("Duplicate rows:", df.duplicated().sum())
```

```python
# Har column ka data type dekho
print("Data types:")
print(df.dtypes)
```

```python
# Timezone hatao date se
df.index = df.index.tz_localize(None)

# Confirm karo
print("Date fixed:")
print(df.index[:10])

# Baad mein jab hum ML model banayenge toh date se naye features nikalenge jaise month, year, day of week etc. Agar timezone laga hota toh wahan error aata! Isliye abhi hi fix kar liya. ✅
```

```python
# Unnecessary columns drop karo
df = df.drop(columns=["Dividends", "Stock Splits"])

# Confirm karo
print("Columns after dropping:")
print(df.columns)
print("New shape:", df.shape)
```

```python
df.head()
```

```python
# Poore data ka summary dekho
print("Dataset Info:")
print("=============")
print(f"Total Trading Days: {df.shape[0]}")
print(f"Total Columns: {df.shape[1]}")
print(f"Date Range: {df.index[0].date()} to {df.index[-1].date()}")
print(f"\nBasic Statistics:")
df.describe()



# Min price ₹395 — ye COVID-19 crash (March 2020) ka time tha jab pure market ne gira tha! Aur tab se price 4x ho gayi ₹1592 tak! Ye ek powerful business insight hai jo tum resume mein mention kar sakte ho! 😎
```

```python
# Libraries import karo
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots

# Data load karo
ticker = yf.Ticker("RELIANCE.NS")
df = ticker.history(start="2020-01-01", end="2026-04-05")

# Clean karo
df.index = df.index.tz_localize(None)
df = df.drop(columns=["Dividends", "Stock Splits"])
df = df[df['Volume'] > 0]

print("Data loaded successfully! ✅")
print("Shape:", df.shape)
```

```python
# Plotly Jupyter ke saath kaam kare isliye ye install karo
!pip install nbformat ipywidgets
```

```python
import plotly.io as pio
pio.renderers.default = 'iframe'

# Line chart
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df.index,
    y=df['Close'],
    mode='lines',
    name='Close Price',
    line=dict(color='#2196F3', width=2)
))

fig.update_layout(
    title='Reliance Industries - Stock Price Journey (2020-2026)',
    xaxis_title='Date',
    yaxis_title='Price (₹)',
    template='plotly_dark',
    hovermode='x unified',
    height=500
)

fig.show()
```

```python
pio.renderers.default = 'iframe'
```

```python
# Candlestick chart
fig2 = go.Figure(data=[go.Candlestick(
    x=df.index,
    open=df['Open'],
    high=df['High'],
    low=df['Low'],
    close=df['Close'],
    increasing_line_color='#26a69a',   # Green = price badhi
    decreasing_line_color='#ef5350'    # Red = price ghati
)])

fig2.update_layout(
    title='Reliance Industries - Candlestick Chart (2020-2026)',
    xaxis_title='Date',
    yaxis_title='Price (₹)',
    template='plotly_dark',
    height=500,
    xaxis_rangeslider_visible=True
)

fig2.show()
```

```python
# browser me run karwane ke liye ek html file banayi hai/ ya bana rahe hai yahan
fig.write_html("reliance_line_chart.html")
print("Chart saved! ✅")
```

```python
pio.renderers.default = 'notebook_connected'
```

```python
# Volume chart - better colors
fig3 = go.Figure()

fig3.add_trace(go.Bar(
    x=df.index,
    y=df['Volume'],
    name='Volume',
    marker_color='#7C83FD'
))

fig3.update_layout(
    title='Reliance Industries - Daily Trading Volume (2020-2026)',
    xaxis_title='Date',
    yaxis_title='Volume (Shares)',
    template='plotly_dark',
    height=400
)

fig3.show()
```

```python
# Daily returns calculate karo
df['Daily_Return'] = df['Close'].pct_change() * 100

# Distribution chart
fig4 = go.Figure()

fig4.add_trace(go.Histogram(
    x=df['Daily_Return'].dropna(),
    nbinsx=100,
    name='Daily Returns',
    marker_color='#7C83FD',
    opacity=0.8
))

fig4.add_vline(
    x=0,
    line_dash="dash",
    line_color="white",
    annotation_text="Zero Line"
)

fig4.update_layout(
    title='Reliance Industries - Daily Returns Distribution (2020-2026)',
    xaxis_title='Daily Return (%)',
    yaxis_title='Frequency (Number of Days)',
    template='plotly_dark',
    height=400
)

fig4.show()
```
